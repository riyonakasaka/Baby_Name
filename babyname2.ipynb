```python
import pandas as pd

dfs = []
years = range(1980, 2024) 
for year in years:
    filename = f"yob{year}.txt"
    with open(filename, "r") as f:
        df = pd.read_csv(f, header=None, names=['name', 'sex', 'number'])
        df['year'] = year
        df['ranking'] = df.index + 1  
        dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)
olivia_df = combined_df[(combined_df['name'] == 'Olivia') & (combined_df['sex'] == 'F')]
olivia_df = olivia_df[['year', 'ranking', 'number']]

olivia_df.to_csv('olivia.csv')
olivia_df
```

```python
dfs = []
years = range(1980, 2024)  
for year in years:
    filename = f"yob{year}.txt"
    with open(filename, "r") as f:
        df = pd.read_csv(f, header=None, names=['name', 'sex', 'number'])
        df['year'] = year
        
        df_top_20 = df.nlargest(20, 'number')
        count_a_ending_names = df_top_20[df_top_20['name'].str.endswith('a')].shape[0]
        dfs.append({'year': year, 'count_a_ending_names': count_a_ending_names})

result_df = pd.DataFrame(dfs)
result_df.to_csv('ending_a.csv')
```
